```python
# @title Teks judul default
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Generate dummy data
X, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.0, random_state=42)

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train model
model = KMeans(n_clusters=3, random_state=42)
y_kmeans = model.fit_predict(X_scaled)

# Plot
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=y_kmeans, palette="coolwarm", s=60)
plt.title("K-Means Clustering (Dummy Data)")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.legend(title="Cluster")
plt.show()
```
